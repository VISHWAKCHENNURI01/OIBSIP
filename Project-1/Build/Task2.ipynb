```python
import pandas as pd

# Load your data
file_path = r'C:\Users\vishw\Downloads\archive (2)\menu.csv'  # Raw string notation to handle backslashes
print(f"Attempting to load file from: {file_path}")

try:
    df = pd.read_csv(file_path)
    # Calculate basic statistics, ignoring non-numeric columns
    mean_values = df.select_dtypes(include='number').mean()
    median_values = df.select_dtypes(include='number').median()
    mode_values = df.select_dtypes(include='number').mode().iloc[0]
    std_dev_values = df.select_dtypes(include='number').std()

    print("Mean values:\n", mean_values)
    print("Median values:\n", median_values)
    print("Mode values:\n", mode_values)
    print("Standard Deviation values:\n", std_dev_values)
except FileNotFoundError:
    print(f"File not found: {file_path}. Please check the file path and try again.")
except Exception as e:
    print(f"An error occurred: {e}")




```
